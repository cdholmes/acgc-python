# Improve figure appearance with ACGC

This notebook demonstrates how to load and use the acgc figure style, which produces figures that are closer to publication ready than the defaults within Matplotlib.

## Default Maplotlib style

The first figure is an example of the default Matplotlib style.

````python
# Figure showing the default Matplotlib style
import matplotlib.pyplot as plt

f = plt.figure(figsize=(3,3))
ax = f.add_subplot()
ax.scatter( range(10), range(10) )
ax.scatter( range(10), range(10,0,-1) )
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax.set_title('Default Matplotlib style')
````

## ACGC figure style
The next example uses ```acgc.figstyle``` to make figures closer to publication ready. The features include 
* font update (similar to Helvetica)
* minor tick lines on both axes
* spacing adjustments between figure panels
* grid lines optional

 The style is activated by importing the module:
```from acgc import figstyle```

By default, the ACGC style includes grid lines. Grid lines can be turned off with

```figstyle.grid_off()```

Grid lines can be turned on with

```figstyle.grid_on()```  


````python
# Load the style
from acgc import figstyle
f = plt.figure(figsize=(5,3))

# Left panel with gridlines (default)
ax = f.add_subplot(1,2,1)
ax.scatter( range(10), range(10) )
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('ACGC style, grid on')

# Turn gridlines off
figstyle.grid_off()

# Right panel without gridlines
ax = f.add_subplot(1,2,2)
ax.scatter( range(10), range(10) )
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('ACGC style, grid off')
````
